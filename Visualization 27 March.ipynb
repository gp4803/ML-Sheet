# Visualization (Seaborn)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cars.csv')
```

# Line plot

```python
# Line plot
```

```python
x = np.arange(10)
y = np.sqrt(x)
```

```python
plt.figure()
sns.lineplot(x,y)
plt.show()
```

# Histogram(Distplot)

```python
plt.figure()
sns.distplot(df['price'])
plt.show()
```

```python
plt.figure()
sns.distplot(df['width'],bins = 20)
plt.show()
```

# Count plot

```python
plt.figure()
sns.countplot(data = df,x = 'body-style',hue='fuel-type')
plt.show()
```

```python
plt.figure()
sns.countplot(data = df,y = 'body-style')
plt.show()
```

# Scatter plot

```python
plt.figure()
sns.scatterplot(data = df,x = 'width',y = 'price')
plt.show()
```

```python
plt.figure()
sns.scatterplot(data = df,x = 'width',y = 'price',hue = 'fuel-type')
plt.show()
```

# Box plot

```python
plt.figure()
sns.boxplot(data = df,x = 'price')
plt.show()
```

```python
plt.figure()
sns.boxplot(data = df,x = 'price',y = 'make')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data = df,x = 'price',y = 'make')
plt.show()
```

# Subplot

```python
x = [1,2,3,4]
y = [5,6,7,8]

fig, axes = plt.subplots(2,2,figsize=(7,7))
axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,0].plot(x,y)
axes[1,1].scatter(x,y)
plt.show()
```

```python
fig, axes = plt.subplots(2,2,figsize=(20,15))
sns.distplot(df['price'],ax = axes[0,0])
sns.countplot(data = df,y = 'make',ax = axes[0,1])
sns.scatterplot(data = df,x = 'width',y = 'price',hue = 'body-style',ax = axes[1,0])
sns.boxplot(data = df,x = 'price',ax = axes[1,1])
plt.show()
```

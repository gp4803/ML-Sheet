# Figsize,labels and Title

```python
area = [600,800,1000,1200,1500]
price = [15,18,25,35,40]
```

```python
plt.figure(figsize = (6,4))
plt.title('Area vs price Plot')

plt.xlabel('Area')
plt.ylabel('Price')

plt.plot(area,price)
plt.show()
```

# xlim and ylim to define x and y range

```python
plt.figure(figsize = (6,4))
plt.title('Area vs price Plot')

plt.xlim((500,2000))
plt.ylim((10,50))

plt.xlabel('Area')
plt.ylabel('Price')

plt.plot(area,price)
plt.show()
```

# xticks and yticks to rotate labels

```python
plt.figure(figsize = (6,4))
plt.title('Area vs price Plot')

plt.xticks(rotation=45)
plt.yticks(rotation=90)

plt.xlabel('Area')
plt.ylabel('Price')

plt.plot(area,price)
plt.show()
```

# Multiple Plot

```python
x = np.arange(1,20)

y1 = np.sqrt(x)
y2 = np.log(x)
```

```python
plt.figure()

plt.plot(x,y1,color = 'blue')
plt.plot(x,y2,color = 'yellow')

plt.legend(labels= ['sqrt','log'])

plt.show()
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure()
plt.hist(df['width'])

plt.title('width frequency dist')
plt.xlabel('width')
plt.ylabel('frequency')

plt.show()
```

```python
df['width'].hist()
plt.show()
```

```python
plt.figure()
plt.hist(df['width'])

plt.title('width frequency dist')
plt.xlabel('width')
plt.ylabel('frequency')

plt.axvline(df['width'].mean(),color = 'red')
plt.axvline(df['width'].median(),color = 'yellow')

plt.show()
```

# Bar Plot

```python
df['make'].unique()
```

```python
df['make'].value_counts()
```

```python
plt.figure(figsize=(7,7))
df['make'].value_counts().plot(kind ='bar')
plt.show()
```

```python
plt.figure(figsize=(7,7))
df['make'].value_counts().sort_values(ascending=False).plot(kind ='bar')
plt.show()
```

```python
plt.figure(figsize=(7,7))
df['make'].value_counts().sort_values(ascending=True).plot(kind ='bar')
plt.show()
```

```python
plt.figure(figsize=(7,7))
df['make'].value_counts().plot(kind ='barh')
plt.show()
```
